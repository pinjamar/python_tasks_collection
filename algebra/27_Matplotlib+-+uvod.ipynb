## Vizualizacija unutar Pandasa

Pandas za oba tipa objekata (Series i DataFrames) ima ugrađenu .plot() metodu. Pandas .plot() metoda u pozadini koristi matplotlib modul. 

Neki od najčešće korištenih argumenata Pandas .plot() metode:
- kind - definira tip grafa (string)
- x ili y - definira jednu ili više kolona koje će biti na nekoj osi (string)
- subplots - Boolean - uključuje li ili ne uključuje dodatne grafove
- layout - pomoću redova, kolona definira broj dodatnih grafova
- figsize - definira veličinu okvirs (širina, visina)
- title - naslov grafa
- grid - uključuje ili ne uključuje linije mreže (Boolean)
- color - pomoću rgb definira boju grafa (red, green, blue)

Problem je što imamo kolonu s datumima koju matplotlib želi pretvoriti u svoj format datuma. Matplotlib ima klasu mdates koja ima metodu date2num, a ona datum pretvara u matplotlib format vremena. 

Prvo moramo napraviti import mdates klase iz matplotlib modula i kolonu pretvcoriti u prikladan format

```python
import matplotlib.dates as mdates
df['dateRep'] = mdates.date2num(df['dateRep'])
```

```python
df.dtypes
```

```python
df.plot()
```

Napisat ćemno da je graf linijski tip, os X će biti kolona "dateRep", a Y kolona će biti "cases". Veličina okvira neka bude širine 10, a visine 5. Uključit ćemo legendu, a naslov će biti "Broj Covid Slučaja". Dodat ćemo i style opciju pomoću koje ćemo podesiti tip i boju linije (puna linija, plave boje)

```python
df.plot(
    kind = 'line',
    x = 'dateRep',
    y = 'cases',
    figsize= (10, 5),
    style = '-b',
    legend = True,
    title = 'Broj Covid Slučajeva'
)
```

```python
df.groupby('dateRep', as_index = False).mean().plot(
    kind = 'line',
    x = 'dateRep',
    y = 'cases',
    figsize= (10, 5),
    style = '-b',
    legend = True,
    title = 'Broj Covid Slučajeva'
)
```

```python
df.groupby('dateRep', as_index = False).mean().plot(
    kind = 'line',
    x = 'dateRep',
    y = 'cases',
    xlabel = 'Datum',
    ylabel = 'Broj zaraženih',
    figsize= (10, 5),
#     style = '-b',
    color = 'green',
    linestyle = 'dashdot',
    legend = True,
    title = 'Broj Covid Slučajeva'
)
```

```python
df.groupby('dateRep', as_index = False).mean().plot(
    kind = 'line',
    x = 'dateRep',
    y = ['cases', 'deaths'],
    xlabel = 'Datum',
    ylabel = 'Broj zaraženih / umrlih',
    figsize= (10, 5),
#     style = '-b',
    color = ['green', 'red'],
    linestyle = 'solid',
    legend = True,
    title = 'Broj Covid Slučajeva'
)
```

Broj smrtnih slučajeva u odnosu na broj zaraženih je vrlo malen, gotovo se ne vidi na grafu. Ajmo dohvatiti posljednjih 40 redaka. 

```python
df.groupby('dateRep', as_index = False).mean().tail(40).plot(
    kind = 'line',
    x = 'dateRep',
    y = ['cases', 'deaths'],
    xlabel = 'Datum',
    ylabel = 'Broj zaraženih / umrlih',
    figsize= (10, 5),
#     style = '-b',
    color = ['green', 'red'],
    linestyle = 'solid',
    legend = True,
    title = 'Broj Covid Slučajeva'
)
```

Izbacimo broj slučajeva iz grafa i prikažimo samo broj umrlih

```python
df.groupby('dateRep', as_index = False).mean().tail(40).plot(
    kind = 'line',
    x = 'dateRep',
    y = 'deaths',
    xlabel = 'Datum',
    ylabel = 'Broj umrlih',
    figsize= (10, 5),
#     style = '-b',
    color = 'red',
    linestyle = 'solid',
    legend = True,
    title = 'Broj Smrti uzrokovanih virusom COVID 19'
)
```

Ovo je odlična primjer za uključiti više grafova u jedan okvir. Pandas nam omogućuje jednostavno dodavanje više grafova unutar jednog okvira. Potrebno je samo uključiti "subplots" i definirati raspored, odnosno layout. 

```python
df.groupby('dateRep', as_index = False).mean().tail(40).plot(
    kind = 'line',
    x = 'dateRep',
    y = ['cases', 'deaths'],
    xlabel = 'Datum',
    ylabel = 'Broj slučajeva',
    figsize = (10, 5),
    subplots = True,
    layout = (1, 2),
    linestyle = 'solid',
    title = 'Prikaz COVID 19 slučajeva i smrti'
)
```

Kreirat ćemo pivot tablicu s jednom dodatnom metodom rolling() koja računa srednju vrijednost za grupe podataka definirane pomoću argumenta. Uzima podatke u zadanom okviru (3, 5, 7 ... dana) izračuna srednju vrijednost za te podatke i onda uzima sljedeću prozor i računa srednju vrijednost za te podatke, itd.

```python
novi_slucajevi_prosjek = df.pivot_table(
    index = df.index, 
    columns = 'countriesAndTerritories', 
    values = 'cases').rolling(7).mean()
```

```python
novi_slucajevi_prosjek
```

```python
okvir, axes = plt.subplots(1, 3, figsize = (15, 5))

novi_slucajevi_prosjek[['Germany']].plot(ax = axes[1], style = '-.c')
novi_slucajevi_prosjek[['France']].plot(ax = axes[0], style = '-.c', title = '7-dnevni prosjek novih COVID-19 slučajeva')
novi_slucajevi_prosjek[['Croatia']].plot(ax = axes[2], style = '-.c')
```

```python
earthquakes_df = pd.read_csv('earthquakes.csv')
earthquakes_df.head(5)
```

```python
earthquakes_df.columns
```

```python
earthquakes_df[['mag', 'magType']].groupby('magType').boxplot(figsize = (15, 8), subplots = False)

plt.title('Magnitude potresa po tipu magnituda')
plt.ylabel('Magnituda')
```

Prikazat ćemo "bar" tip podatka. Koristimo ga za prikaz broja ponavljanja vrijednosti u nekoj varijabli (count)

```python
earthquakes_df.place.value_counts().iloc[15::-1].plot(
    kind = 'barh',
    figsize = (10, 5)
)
```

```python
earthquakes_df.groupby('place').tsunami.sum().sort_values().iloc[-15:].plot(
    kind = 'barh',
    figsize = (10, 5)
)
```

```python
earthquakes_df.place.head(50)
```

```python
earthquakes_df['mjesta'] = earthquakes_df.place.str.replace(
    r'.* of ','', regex = True
).str.replace(
    'the ', ''
).str.replace(
    'NV', 'Nevada'
).str.replace(
    'CA', 'California'
).str.replace(
    r' region', '', regex = True
).str.replace(
    'MX', 'Mexico'
).str.replace(
    'Fiji Islands', 'Fiji'
).str.replace(
    r'.*, ', '', regex = True
).str.strip()
```

```python
earthquakes_df['mjesta'].sort_values().unique()
```

```python
earthquakes_df.groupby('mjesta').tsunami.sum().sort_values().iloc[-10:].plot(
    kind = 'barh',
    figsize = (10, 5),
    title = '10 mjesta s najviše tsunamija'
)
```

```python
indonesia_earthquakes = earthquakes_df.query(
    'mjesta == "Indonesia"'
).assign( # iz filtriranih podataka kreiraj novi objekt koji će imat izmijenjene podatke tako što
    time = lambda x: pd.to_datetime(x.time, unit= 'ms'), # će svaki datum unutar time kolone pretvoriti u Pandas vrijeme
    earthquake = 1
).set_index('time').resample('1D').sum()
```

```python
indonesia_earthquakes.index = indonesia_earthquakes.index.strftime('%b\n%d')
```

```python
indonesia_earthquakes.plot(
    kind = 'bar',
    rot = 0,
    y = ['earthquake', 'tsunami'],
    figsize = (15, 3),
    label = ['Potresi', 'Tsunamiji'],
    title = 'Broj potresa i tsunamija u Indoneziji po danu'
)
```

Zanima nas koliko često se nakon potresa pojaci tsunami na pojedinoj lokaciji. 

```python
earthquakes_df.groupby(['mjesta', 'tsunami']).mag.count().unstack().apply(
    lambda x: x/x.sum(), axis = 1
).rename(
    columns = {0: 'Ne', 1: 'Da'}
).sort_values('Da', ascending = False)[7::-1].plot.barh()
```
